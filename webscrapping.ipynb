```python
# importing libraries
import pandas as pd
import requests as re
from bs4 import BeautifulSoup
```

```python
# scrapping through website
page=re.get("https://www.worldometers.info/world-population/population-by-country/")
```

```python
# Extracting data with the help of beautifulsoup
sp=BeautifulSoup(page.content)
```

```python
table=sp.find_all("td")
```

```python
# using for loop to add all the data in empty list 
list=[]
for i in table:
 list.append(i.text)
list
```

```python
list1=list[0:len(list):12]
list2=list[1:len(list):12]
list3=list[2:len(list):12]
list4=list[3:len(list):12]
list5=list[4:len(list):12]
list6=list[5:len(list):12]
list7=list[6:len(list):12]
list8=list[7:len(list):12]
list9=list[8:len(list):12]
list10=list[9:len(list):12]
list11=list[10:len(list):12]
list12=list[11:len(list):12]
print(len(list1),len(list2),len(list3),len(list4),len(list5),len(list6),len(list7),len(list8),len(list9),len(list10),len(list11),len(list12))
```

```python
# Creating Dataframe
df=pd.DataFrame({})
df["SR NO"]=list1
df["COUNTRY"]=list2
df["POPULATION"]=list3
df["YEARLY CHANGE"]=list4
df["NET CHANGE"]=list5
df["DENSITY(P/Km)"]=list6
df["LAND AREA(Km)"]=list7
df["MIGRANTS"]=list8
df["FERT RATE"]=list9
df["MED AGE"]=list10
df["URBAN POP"]=list11
df["WORLD SHARE"]=list12
df
```

```python
# converting data into CSV
df.to_csv(r"C:\Users\parth\OneDrive\Desktop\parth\web1.csv")
```
